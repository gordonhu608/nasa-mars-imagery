## Generate TFRecords

```python
display.display(display.Image(filename=df[0][0]))
```

```python
def _bytes_feature(value):
    return tf.train.Feature(bytes_list=tf.train.BytesList(value=[value]))

def _float_feature(value):
    return tf.train.Feature(float_list=tf.train.FloatList(value=[value]))

def _int64_feature(value):
    return tf.train.Feature(int64_list=tf.train.Int64List(value=[value]))
```

```python
def create_tf_example(filename, label):
    image = cv2.imread(filename)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    image = cv2.resize(image, (RESIZE, RESIZE))
    
    feature = {
        'image': _bytes_feature(image.tobytes()),          
        'label': _int64_feature(label),
    }
    tf_example = tf.train.Example(features=tf.train.Features(feature=feature))
    return tf_example
```

```python
RESIZE = 64
df = readDatasetDF('train')
writer = tf.io.TFRecordWriter("data/out/train.tfrecords")
    
for i, img_info in df.iterrows():
    filename = img_info[0]
    label = img_info[1]
    tf_example = create_tf_example(filename, label)
    writer.write(tf_example.SerializeToString())
```

```python
def _extract_fn(tfrecord):
    feature = {
        'image': tf.io.FixedLenFeature([], tf.string),
        'label': tf.io.FixedLenFeature([], tf.int64),
    }
    sample = tf.io.parse_single_example(tfrecord, feature)
    
    image = tf.io.decode_raw(sample['image'], tf.uint8)
    image = image / 64
    image = tf.reshape(image, [RESIZE, RESIZE, 1])
    image = tf.dtypes.cast(image, tf.float32)
    
    label = tf.cast(sample['label'], tf.int64)
    
    return [image, label]
```

```python
def batch_dataset(dataset, batch_size):
    dataset = dataset.map(_extract_fn)
    dataset = dataset.batch(batch_size, drop_remainder=True)
    return dataset
```

```python
BATCH_SIZE = 64

train_tfr = tf.data.TFRecordDataset('data/out/train.tfrecords')
train_ds = batch_dataset(train_tfr, BATCH_SIZE)
```

## Transfer Learning

```python
# Baseline Resnet50 model
base_model = tf.keras.applications.ResNet50(
    include_top=False, weights=None, input_tensor=None, 
    input_shape=(64,64,1), pooling=None
)
```

```python
base_model.trainable = False
```

```python
base_model.summary()
```

```python
inputs = tf.keras.Input(shape=(64,64,1))
x = base_model(inputs, training=False)

#Averaging layer to convert features into 2048-element vector per image
global_average_layer = tf.keras.layers.GlobalAveragePooling2D()
x = global_average_layer(x)

#Dense layer to convert features into single prediction 
prediction_layer = tf.keras.layers.Dense(1)
outputs = prediction_layer(x)

model = tf.keras.Model(inputs, outputs)
```

```python
# Compile the Model
base_learning_rate = 0.0001
model.compile(optimizer=tf.keras.optimizers.Adam(lr=base_learning_rate),
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False),
              metrics=['accuracy'])
```

```python
initial_epochs = 10
history = model.fit(train_ds, epochs=initial_epochs)
```

```python
# Subset the data
df = df.sample(frac=1)

l = []
for i in df[1].unique():
    temp = df.loc[df[1]==i]
    l.append(temp.sample(frac=1)[:4])

df_sub = pd.concat(l) 
```
